### Training ICC Single branch and SRGAN pipeline

Epochs - 320

Batch size = 2


#### ICC Network (weights pre-initialized)

Learning Rate = 0.00001

Momentum = 0.9

Loss Function = 1000.0 * MSELoss(GT_Density, Density)

Optimizer = Adam


#### SRGAN

Alpha = 2

Generator Loss Function = 0.01 \* (image_loss + 0.001 \* adversarial_loss + 0.006 \* perception_loss) + density_loss

where, image_loss is MSE Loss between new image and original image

perception_loss is MSE Loss for VGG features

Optimizer = Adam

#### Best Results

MAE_HR: 88.1

MSE_HR: 138.5

Epoch: 225

Code - https://github.com/udbhav-sharma/Crowd_SRGAN/tree/srgan_icc_together
